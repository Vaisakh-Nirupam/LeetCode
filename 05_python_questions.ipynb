Q1. Find out Common Letters 

```python
str1 = set(input("Enter First String: "))
str2 = set(input("Enter Second String: "))
def common(s1,s2):
    return s1 & s2
print(common(str1,str2))

# or

common = lambda s1,s2 : s1 & s2
print(common(str1,str2))
```

```python
lst = [5,7,4,3,9,8,19,21,12]
tar = 17
dit = {}
for i,num in enumerate(lst):
    sub = tar - num
    if sub in dit:
        print(f"{dit[sub]},{i}")
    dit[num] = i
```

Q6. Minimum Difference bw each Elements

```python
lst = [5,32,45,4,12,18,25]
print("List:",lst)
lst.sort()
print("Sorted List:",lst)

min_val = float('inf')
for i in range(len(lst)-1):
    diff = lst[i+1] - lst[i] 
    if diff < min_val:
        min_val = diff
print("Minimum Difference:",min_val)

# or

min_diff = min(b - a for a, b in zip(lst, lst[1:]))
print("Minimum Difference:", min_diff)
```

Q7. Maximum Difference bw each Elements

```python
lst = [5,32,45,4,12,18,25]
print("List:",lst)
lst.sort()
print("Sorted List:",lst)
max_val = float('-inf')
for i in range(len(lst)-1):
    diff = lst[i+1] - lst[i] 
    if diff > max_val:
        max_val = diff
print("Maximum Difference:",max_val)

# or

max_diff = max(b - a for a, b in zip(lst, lst[1:]))
print("Maximum Difference:", max_diff)
```

Q8. Stack Push, Pop and Operate

```python
def stack_operate(lst):
    op = ["+","-","*","/","%"]
    stk = [] 
    for i in lst:
        if i not in op:
            stk.append(int(i))
        else:
            n1, n2 = stk.pop(), stk.pop()
            if i == '+':
                stk.append(n1+n2)
            elif i == '-':
                stk.append(n1-n2)
            elif i == '*':
                stk.append(n1*n2)
            elif i == '/':
                stk.append(n1/n2)
            elif i == '%':
                stk.append(n1%n2)
    return stk[-1]
print(stack_operate(["1","2","+","3","*"]))
```

Q9. Create Wave Array from Normal Arrays

```python
def waves(arr):
    for i in range(0,len(arr)-1,2):
        if arr[i] < arr[i+1]:
            arr[i],arr[i+1] = arr[i+1],arr[i]
    return arr
arr = [3,5,12,2,6,10,7,9,8]
print("Before Wave:",arr)
print("After Wave:",waves(arr))
```

Q10. Find the Indexes of the Anagrams

```python
def anagram(lst):
    dit = {}
    for i in range(len(lst)):
        word = "".join(sorted(lst[i]))
        if word in dit:
            dit[word].append(i)
        else:
            dit[word] = [i]
    return dit 
lst = ["cat","god","dog","act","tac"]
print(anagram(lst))
```

Q11. Check all the possible Rotations

```python
def all_rotations(s):
    if s is None:
        return []
    rotations = []
    for i in range(len(s)):
        rotations.append(s[i:] + s[:i])
    return rotations
print(all_rotations("abcd"))
```

Q12. Check if its a Rotation

```python
def rotated(a,b):
    if a and b == None:
        return False
    if len(a) != len(b):
        return False
    return b in a+a
A = "abcd"
B = "dabc"
print(rotated(A,B))
```

Q13. First Non Repeating Letter

```python
def nonrep(word):
    dit = {}
    for i,j in enumerate(word):
        if j in dit:
            dit[j][1] += 1
        else:
            dit[j] = [i,1]
    for i in dit:
        if dit[i][1] == 1:
            return dit[i][0]
    return None
word = "EXCELLENT"
print(nonrep(word)) 
```

Q15. Sort Balls in Order Without Sort

```python
def sort_balls(lst):
    red = 0
    blue = 0
    all = len(lst)
    for i in range(all):
        if lst[i] == 1:
            red+=1
        elif lst[i] == 2:
            blue+=1
    return [1]*red + [2]*blue + [3]*(all-red-blue)
lst = [1,2,3,1,3,2,1,2,3,1]
print(sort_balls(lst))
```

Q16. Prime or Not

```python
def prime(num):
    c=0
    for i in range(1,num+1):
        if num % i == 0:
            c+=1
    if c == 2:
        return "prime"
    return "not a prime"
print(prime(7))
# or
def prime(num):
    if num <= 1:
        return "not prime"
    for i in range(2,int(num**0.5)+1):
        if num % i == 0:
            return "not prime"
    return "prime"
print(prime(7))
```

Q18.  Fibonacci Series

```python
def fibo(num):
    a,b = 0,1
    f = []
    for _ in range(num):
        f.append(a)
        a,b=b,a+b
    return f
print(fibo(5))
```

Q19. Strong Password

```python
import re
def pass_check(pw):
    pattern = r"^(?=.*[a-z])(?=.*[A-Z])(?=.*[0-9])(?=.*[!@#$%^&*()]).{8,}$"
    if re.match(pattern,pw):
        return "Valid Password"
    return "Invalid Password"

print(pass_check("Password1!"))
print(pass_check("password1"))
```
